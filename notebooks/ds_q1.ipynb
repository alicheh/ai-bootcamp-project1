```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv('../data/HBN.csv')
data.head()
```

```python

sns.set(style="whitegrid")
plt.figure(figsize=(15, 20))

#Age
plt.subplot(3,2,1)
sns.histplot(data['Age'], kde=True, color='blue', bins=15)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

#Weight
plt.subplot(3,2,2)
sns.histplot(data['Physical-Weight'], kde=True, color='green', bins=15)
plt.title('Weight Distribution')
plt.xlabel('Weight (lbs)')
plt.ylabel('Count')

# Height
plt.subplot(3,2,3)
sns.histplot(data['Physical-Height'], kde=True, color='orange', bins=15)
plt.title('Height Distribution')
plt.xlabel('Height (Inch)')
plt.ylabel('Count')

# Internet usage
plt.subplot(3,2,4)
sns.countplot(x='PreInt_EduHx-computerinternet_hoursday', data=data, color='purple')
plt.title('Internet Usage Distribution')
plt.xlabel('Internet Usage (hours/day)')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Less than 1h/day','Around 1h/day', 'Around 2hs/day', 'More than 3hs/day'])

#Gender disturbution
plt.subplot(3,2,5)
sns.countplot(x='Sex', data=data, color = 'red')
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Men', 'Women'])


plt.tight_layout()
plt.show()
```
