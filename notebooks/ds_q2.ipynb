```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.cluster import DBSCAN
```

```python
data = pd.read_csv('../data/HBN.csv')
data.head()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.boxplot(x='Sex', y='BIA-BIA_BMI', data=data)
plt.title('BMI/Sex')
plt.xlabel('Sex')
plt.ylabel('BMI')

plt.subplot(1, 3, 2)
sns.boxplot(x='Sex', y='BIA-BIA_BMR', data=data)
plt.title('BMR/Sex')
plt.xlabel('Sex')
plt.ylabel('BMR')

plt.subplot(1, 3, 3)
sns.boxplot(x='Sex', y='CGAS-Score', data=data)
plt.title('CGAS/Sex')
plt.xlabel('Sex')
plt.ylabel('CGAS-Score')

plt.tight_layout()
plt.show()
```
